```python
import pandas as pd

df = pd.read_csv('olympics.csv')
df.head()
```

```python
df.dropna(subset=['medal'], inplace=True)
df.head()
df.info()
```

```python

medal_counts = df.groupby('noc').size().reset_index(name='total_medals')

# Filter countries with more than 1000 medals
countries_with_1000_medals = medal_counts[medal_counts['total_medals'] >= 1000]['noc']

# Keep only the entries for countries with more than 1000 medals
df_filtered = df[df['noc'].isin(countries_with_1000_medals)]

countries_with_1000_medals.head()
df_filtered.to_csv('clean_olympics.csv', index=False)
```
